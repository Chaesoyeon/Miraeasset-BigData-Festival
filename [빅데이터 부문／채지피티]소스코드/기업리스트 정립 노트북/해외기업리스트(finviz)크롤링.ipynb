```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.common.keys import Keys
```

```python
#자동 업데이트
from webdriver_manager.chrome import ChromeDriverManager

#꺼짐 방지
chrome_options = Options()
chrome_options.add_experimental_option('detach', True)

#에러 없애기
chrome_options.add_experimental_option('excludeSwitches',['enable-logging'])
service = Service(executable_path=ChromeDriverManager().install())
driver=webdriver.Chrome(service=service, options = chrome_options)
driver.implicitly_wait(5)
driver.maximize_window()

item_list = []
lastpage = 6  #페이지 바꾸기
for page in range(1,lastpage*20,20):                            #섹터 바꾸기
    driver.get(f'https://finviz.com/screener.ashx?v=111&f=sec_utilities&r={page}')
    time.sleep(2)
    items = driver.find_elements(By.CLASS_NAME, "screener-body-table-nw")
    for item in items:
        print(item)
    
```

```python
data = []
row_count = len(item_list)//11
col_count = 11

for i in range(row_count):
    row_data = item_list[i * col_count:(i + 1) * col_count]
    data.append(row_data)

df = pd.DataFrame(data).iloc[:,1:11]
df.columns = ['Ticker','Company','Sector','Industrials','Country','Market Cap','P/E','Price','Change','Volume']
df = df[df['Country']=='USA'].reset_index(drop=True)
```
