```python
import pandas as pd
import FinanceDataReader
```

```python
A = pd.read_csv('국내 주식 소분류.csv', encoding = 'cp949')
B = pd.read_csv('국내 주식 소분류_unique.csv', encoding = 'cp949')
```

```python
def change_dataset(file1, file2):
    for idx, row in file1.iterrows():
        if pd.notnull(row['변경']):
            mask = file2['기업'] == row['기업']
            file2.loc[mask, '소분류'] = row['변경']
        
    return file2
```

```python
change_df = change_dataset(B, A)
change_df
```

## 소분류를 공통된 산업분류로 재분류 작업 진행

### 채소연 part_1-83

```python
c = pd.read_excel('섹터분류_중간_소연.xlsx')
sy = c.iloc[:83, 4:6]
sy
```

```python
def update_sector(original_df, update_file):
    for idx, row in update_file.iterrows():
        target_value = row['소분류']
        updated_value = row['Unnamed: 5']
        
        original_df.loc[original_df['소분류'] == target_value, '소분류'] = updated_value
        
    return original_df
```

```python
update_df1 = update_sector(change_df, sy)
update_df1
```

### 채희지 part

```python
d = pd.read_excel('섹터분류_중간_희지.xlsx')
hj = d.iloc[83:165, 4:6]
hj
```

```python
update_df2 = update_sector(update_df1, hj)
update_df2
```

### 윤지영 part

```python
e = pd.read_excel('섹터분류_윤지영.xlsx')
jy = e.iloc[165:, [4,6]]
jy
```

```python
# 추가 수정 사항
jy.loc[jy['소분류'] == '화학섬유', 'Unnamed: 6'] = 'Specialty Chemicals'
jy = jy.rename(columns = {'Unnamed: 6' : 'Unnamed: 5'})
```

```python
update_df3 = update_sector(update_df2, jy)
update_df3
```

### 해외 섹터 전처리

```python
usa = pd.read_csv('C:/Users/pc/미래에셋/해외_섹터.csv', encoding = 'cp949')
```

#### ETF 종목 제거

```python
new_usa = usa[usa['Industrials'] != 'Exchange Traded Fund']
new_usa.reset_index(drop = True, inplace = True)
```

#### 불필요한 열 제거

```python
columns_to_drop = ['Sector', 'Country', 'Market Cap', 'P/E', 'Change', 'Volume', 'Price']
new_usa = new_usa.drop(columns = columns_to_drop)
new_usa
```

#### 시장 정보 추가

```python
nsd = FinanceDataReader.StockListing('NASDAQ')
amex = FinanceDataReader.StockListing('AMEX')
nyse = FinanceDataReader.StockListing('NYSE')
```

```python
market_dataframe = {'NASDAQ' : nsd, 'AMEX' : amex, 'NYSE' : nyse}
```

```python
new_usa['Market'] = None
for market, market_df in market_dataframe.items():
    matching_tickers = new_usa[new_usa['Ticker'].isin(market_df['Symbol'])]
    new_usa.loc[matching_tickers.index, 'Market'] = market
```

```python
new_usa
```

#### Market열 결측치 처리

```python
# 결측치 행 제거
new_usa = new_usa[~new_usa['Market'].isna()]
new_usa
```

```python
# Market 값이 Na인 행 추출 후 직접 수작업 처리 진행
market_na = new_usa[new_usa['Market'].isna()]
market_na.to_csv('확인필요_해외.csv', index = False, encoding = 'cp949')
```

```python
# 수작업으로 처리한 파일 적용
market_na_check = pd.read_csv('확인필요_해외_처리완.csv', encoding = 'cp949')

# 결측치 행 제거
market_na_check = market_na_check.dropna(subset = ['Market']).reset_index(drop = True)
market_na_check
```

```python
# 최종 데이터셋
result_usa = pd.concat([new_usa, market_na_check], ignore_index = True)
result_usa
```

```python
result_usa.to_csv('C:/Users/pc/미래에셋/해외_산업분류.csv', index = False, encoding = 'cp949')
```
