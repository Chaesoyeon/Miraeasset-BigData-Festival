```python
import pandas as pd

#미국 산업분류 파일 불러오기 
us_df = pd.read_csv('C:/Users/01061.DESKTOP-UR6KUEV/KT_visang/해외_산업분류(최종).csv', index_col=0)
```

```python
#국내 산업분류 파일 불러오기(미국의 산업분류 번역한 항목)
kr_df = pd.read_excel('C:/Users/01061.DESKTOP-UR6KUEV/KT_visang/섹터분류_중간.xlsx', index_col=0)
kr_df = kr_df[['산업분류', 'Industrials']]
kr_df.reset_index(inplace=True)
kr_df.dropna(inplace=True)
kr_df
```

```python
us_df = us_df.merge(kr_df[['Industrials', '산업분류']], on='Industrials', how='left')
us_df.drop(['Company','Industrials','Market'], axis=1, inplace=True)
us_sector_list = us_df.산업분류.unique()
us_sector_list
```

```python
df = pd.read_csv('C:/Users/01061.DESKTOP-UR6KUEV/KT_visang/국내_산업분류(최종2).csv', encoding='cp949')
df = df.merge(kr_df[['Industrials', '산업분류']], on='Industrials', how='left')
df.drop(['Industrials','종목코드','시장'], axis=1, inplace=True)
df
```

```python
#미국 섹터 원-핫 인코딩(one-hot encoing) 진행
for category in us_sector_list:
    us_df[category] = us_df['산업분류'].apply(lambda x: 1 if x == category else 0)
us_df.drop('산업분류', axis=1, inplace=True)
us_df
```

```python
#국내 섹터 원-핫 인코딩(one-hot encoing) 진행
for category in us_sector_list:
    df[category] = df['산업분류'].apply(lambda x: 1 if x == category else 0)
df.drop(['산업분류'], axis=1, inplace=True)
df
```

```python
#두 가지 이상으로 분류된 기업 처리
sumusdf = us_df.groupby('Ticker').sum()
sumusdf.reset_index(inplace=True)
sumusdf
```

```python
#두 가지 이상으로 분류된 기업 처리
df = df.groupby('기업').sum()
df.reset_index(inplace=True)
df
```

```python
df.to_csv('국내_유사도_기본.csv', encoding='cp949', index=False)
sumusdf.to_csv('해외_유사도_기본.csv', encoding='cp949', index=False)
```
